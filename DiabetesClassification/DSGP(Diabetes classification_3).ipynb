```python
diabetes = pd.read_csv('drive/MyDrive/DSGP(Diabetes)/Data/diabetes_binary_health_indicators_BRFSS2015.csv')
```

```python
diabetes
```

```python
X = diabetes[['HighBP', 'HighChol', 'BMI', 'Smoker', 'PhysActivity', 'Fruits', 'Veggies', 'MentHlth', 'DiffWalk', 'Sex', 'Age', 'Income']]
```

```python
X
```

```python
y = diabetes['Diabetes_binary']
```

```python
y
```

```python
print("Class Distribution:\n", y.value_counts())
```

```python
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.3, random_state=42, stratify=y
)
```

```python
smote = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)
```

```python
rf_clf = RandomForestClassifier(random_state=42)
rf_clf.fit(X_train_resampled, y_train_resampled)
rf_predictions = rf_clf.predict(X_test)
```

```python
dt_clf = DecisionTreeClassifier(random_state=42)
dt_clf.fit(X_train_resampled, y_train_resampled)
dt_predictions = dt_clf.predict(X_test)
```

```python
rf_accuracy = accuracy_score(y_test, rf_predictions)
print("\nRandom Forest Accuracy:", rf_accuracy)
print(classification_report(y_test, rf_predictions))
```

```python
dt_accuracy = accuracy_score(y_test, dt_predictions)
print("\nDecision Tree Accuracy:", dt_accuracy)
print(classification_report(y_test, dt_predictions))
```
